```python
f = np.ones(64)
f = f/sum(f)

g = convolve(f,f)
g = g / sum(g)

h = convolve(g,g)
h = h / sum(g)

j = convolve(h,h)
j = j / sum(j)


plt.subplot(2,2,1)
plt.plot(f,'k-')
plt.axis('off')
plt.subplot(2,2,2)
plt.plot(g,'k-')
plt.axis('off')
plt.subplot(2,2,3)
plt.plot(h,'k-')
plt.axis('off')
plt.subplot(2,2,4)
plt.plot(j,'k-')
plt.axis('off')
plt.show()
```
